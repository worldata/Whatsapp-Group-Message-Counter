```python
import re
datum = []
time = []
name = []
message = []

pattern = "Simon"
pattern2 = "Max"

with open('chat.txt',encoding="utf8") as f:
    lines = f.readlines()
    for line in lines[3:]:
        if line[0] == "[" and line[1] != "X" and line[1] != "]" and line[1] != "S":
            datum.append(line.split(",")[0][1:])
            time.append(line.split(" ")[1][:-1])
            message.append(str(line.split(": ")[1:]))
            
            if line.split(" ")[2] == "Max:" or line.split(" ")[2] == "\u200eYou:"or line.split(" ")[2] == "\u200eYou":
                name.append((str(line.split(" ")[2])))

            elif line.split(" ")[2] == "Simon" or line.split(" ")[2] == "\u200e\u200eSimon" or line.split(" ")[2] == "\u200eSimon":
                name.append(str(line.split(" ")[2] + " " + line.split(" ")[3][:-1]))

            elif line.split(" ")[2] == "Hugo" or line.split(" ")[2] == "\u200eHugo":
                name.append(str(line.split(" ")[2] + " " + line.split(" ")[3])+ " " + line.split(" ")[4][:-1])

            else:
                print(line.split(" ")[0])
                print(line.split(" ")[1])
                print(line.split(" ")[2])
                print(line.split(" "))

    
    
   
```

```python
print(len(datum))
print(len(time))
print(len(name))
print(len(message))
```

```python
import pandas as pd

df = pd.DataFrame(
    {'datum': datum,
     'time': time,
     'name': name, 
     'message' : message
    })

df.message = df.message.apply(lambda x: x[1:-2])

df.head(5)

```

```python
df["name"] = df.name.apply(lambda x:"Max:" if x == "\u200eYou" or x == "\u200eYou:" else x)
df["name"] = df["name"].apply(lambda x: "Hugo Hoppe Stinki" if x == "\u200eHugo Hoppe Stink" or x == "Hugo Hoppe Stinki:" else x)
df["name"] = df["name"].apply(lambda x: "Simon Lassahn" if x == "u200eSimon Lassah" or x == "\u200eSimon Lassah" or x == "\u200e\u200eSimon Lassah" else x)
```

```python
df["name"] = df.name.apply(lambda x:"Max" if x == "Max:" else x)
df["name"].value_counts()
```

```python
df["day"] = df.datum.apply(lambda x: str(x[0:2]))
df["month"] = df.datum.apply(lambda x: str(x[3:5]))
df["year"] = df.datum.apply(lambda x: "20" + str(x[-2:]))
df["moye"] = df.month + df.year

```
